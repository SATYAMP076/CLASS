```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
#replacing zero values with the mean of column
data['BMI'] = data['BMI'].replace(0,data['BMI'].mean())

data['BloodPressure'] = data['BloodPressure'].replace(0,data['BloodPressure'].mean())

data['Glucose'] = data['Glucose'].replace(0,data['Glucose'].mean())

data['Insulin'] = data['Insulin'].replace(0,data['Insulin'].mean())

data['SkinThickness'] = data['SkinThickness'].replace(0,data['SkinThickness'].mean())
```

```python
#find the IQR (inter quantile range) to identify outliers

#1st quantile
q1=data.quantile(0.25)
#3rd quantie
q3=data.quantile(0.75)

iqr=q3-q1
```

Outlier dection formula

highe side= Q3+(1.5* IQR)

lower side=Q1-(1.5*IQR)

```python
preg_high=(q3.Pregnancies +(1.5 * iqr.Pregnancies))
preg_high
```

```python
index= np.where(data['Pregnancies']>preg_high)
index
```

```python
data=data.drop(data.index[index])
data.shape
```

```python
data.reset_index()
```

```python
bp_high=(q3.BloodPressure + (1.5*iqr.BloodPressure))
print(bp_high)

index= np.where(data['BloodPressure']>bp_high)
data=data.drop(data.index[index])
data.shape
data.reset_index()
```

```python
sk_high=(q3.SkinThickness + (1.5*iqr.SkinThickness))
print(sk_high)

index= np.where(data['SkinThickness']>sk_high)
data=data.drop(data.index[index])
data.shape
data.reset_index()
```

```python
in_high=(q3.Insulin + (1.5*iqr.Insulin))
print(in_high)

index= np.where(data['Insulin']>in_high)
data=data.drop(data.index[index])
data.shape
data.reset_index()
```

```python
bmi_high=(q3.BMI + (1.5*iqr.BMI))
print(bmi_high)

index= np.where(data['BMI']>bmi_high)
data=data.drop(data.index[index])
data.shape
data.reset_index()
```

```python
dpf_high=(q3.DiabetesPedigreeFunction + (1.5*iqr.DiabetesPedigreeFunction))
print(dpf_high)

index= np.where(data['DiabetesPedigreeFunction']>dpf_high)
data=data.drop(data.index[index])
data.shape
data.reset_index()
```

```python
age_high=(q3.Age+ (1.5*iqr.Age))
print(age_high)

index= np.where(data['Age']>age_high)
data=data.drop(data.index[index])
data.shape
data.reset_index()
```

```python
bp_low=(q1.BloodPressure - (1.5*iqr.BloodPressure))
print(bp_low)

index= np.where(data['BloodPressure']<bp_low)
data=data.drop(data.index[index])
data.shape
data.reset_index()
```

```python
plt.figure(figsize=(20,25))
plotnumber=1

for column in data:
    if plotnumber<=9 : # as there are 9 column in the data
        ax=plt.subplot(3,3,plotnumber)
        sns.distplot(data[column])
        plt.xlabel(column,fontsize=20)
        
    plotnumber+=1
plt.show()
```

```python
x=data.drop(columns=['Outcome'])
y=data['Outcome']
```

```python
plt.figure(figsize=(20,25), facecolor='white')
plotnumber=1

for column in x:
    if plotnumber<=9 : # as there are 9 column in the data
        ax=plt.subplot(3,3,plotnumber)
        sns.stripplot(y,x[column])
        
    plotnumber+=1
plt.tight_layout()
```

```python
scalar= StandardScaler()
x_scaled=scalar.fit_transform(x)
```

```python
x_scaled.shape[1]
```

```python
##finding variance inflation factor in each scaled column i.e x_scaled.shape[1] (1/(1-R2))
```

```python
vif= pd.DataFrame()
vif["vif"]=[variance_inflation_factor(x_scaled,i) for i in range(x_scaled.shape[1])]
vif["Features"]=x.columns

vif
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.25, random_state=335)
```

```python
log_reg=LogisticRegression()
log_reg.fit(x_train,y_train)
```

```python
import pickle
with open ('modelForPrediction.sav','wb') as f:
    pickle.dump(log_reg,f)
    
with open ('sandardScalar.sav','wb') as f:
    pickle.dump(scalar,f)
```

```python
y_pred= log_reg.predict(x_test)
```

```python
y_pred
```

```python
accuracy=accuracy_score(y_test,y_pred)
accuracy
```

```python
conf_mat= confusion_matrix(y_test,y_pred)
conf_mat
```

```python
tp=conf_mat[0][0]
fp=conf_mat[0][1]
fn=conf_mat[1][0]
tn=conf_mat[1][1]
```

```python
accuracy=(tp+tn)/(tp+fp+fn+tn)
accuracy
```

```python
precision=tp/(tp+fp)
precision
```

```python
recall=tp/(tp+fn)
recall
```

```python
F1=2*(recall*precision)/(recall+precision)
F1
```

```python
auc=roc_auc_score(y_test,y_pred)# area under the curve
auc
```
