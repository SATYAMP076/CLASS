# Merging two DataFrame

```python
df1=pd.DataFrame({'A':['A0','A1','A2','A3'],
             'B':['B0','B1','B2','B3'],
             'C':['C0','C1','C2','C3'],
             'D':['D0','D1','D2','D3']},
             index=[0,1,2,3])
```

```python
df2=pd.DataFrame({'A':['A4','A5','A6','A7'],
             'B':['B4','B5','B6','B7'],
             'C':['C4','C5','C6','C7'],
             'D':['D4','D5','D6','D7']},
             index=[4,5,6,7])
```

```python
df3=pd.DataFrame({'A':['A8','A9','A10','A11'],
             'B':['B8','B9','B10','B11'],
             'C':['C8','C9','C10','C11'],
             'D':['D8','D9','D10','D11']},
             index=[8,9,10,11])
```

```python
print("\nThe DataFrame number 1\n",'-'*30,sep='')
print(df1)
```

```python
print("\nThe DataFrame number 2\n",'-'*30,sep='')
print(df2)
```

```python
print("\nThe DataFrame number 3\n",'-'*30,sep='')
print(df3)
```

```python
df_cat=pd.concat([df1,df2,df3], axis=0)
print("\n After concatenation along row\n",'-'*30, sep='')
print(df_cat)
```

```python
df_cat1=pd.concat([df1,df2,df3], axis=1)
print("\n After concatenation along column\n",'-'*30, sep='')
print(df_cat1)
```

```python
df_cat2=df_cat1.fillna(value=0 )
print("\n After filling missing values with zero\n",'-'*60,sep='')
print(df_cat2)
```

```python
df_cat2
```

```python
left=pd.DataFrame({'key':['K0','K1','K2','K3'],
                  'A':['A0','A1','A2','A3'],
                  'B':['B0','B1','B2','B3']})

right=pd.DataFrame({'key':['K0','K1','K2','K3'],
                  'C':['C0','C1','C2','C3'],
                  'D':['D0','D1','D2','D3']})
```

```python
print("\nThe DataFrame 'left'\n",'-'*30,sep='')
print(left)
```

```python
print("\nThe DataFrame 'right'\n",'-'*30,sep='')
print(right)
```

```python
merge1=pd.merge(left,right,how='inner',on='key')

print("\nAfter simple merging with 'inner' method\n",'-'*30,sep='')
print(merge1)
```
